### Задание 2: Комплексное число (2 балла):
Реализуйте свой класс Complex для комплексных чисел, аналогично встроенной реализации complex:
1. Добавьте конструктор класса
2. Реализуйте операции проверки на равенство, сложения, вычитания, произведения и деления комплексных чисел. Используйте магические методы! (__eq__, __add__, __sub__, __mul__, __truediv__)
3. Реализуйте операцию модуля (__abs__)
4. Оба класса должны давать осмысленный вывод как при print, так и просто при вызове в ячейке (погуглите соответсвующие магические методы)

````python
class Complex:
    def __init__(self,real,mnim):
        self.real=real
        self.mnim=mnim
    def __eq__(self,other):
        if (self.real==other.real) and (self.mnim==other.mnim):
            return True
        return False
    def __add__(self,other):
        result=Complex(0,0)
        result.real=self.real+other.real
        result.mnim=self.mnim=other.mnim
        return result
    def __sub__(self,other):
        result=Complex(0,0)
        result.real=self.real-other.real
        result.mnim=self.mnim-other.mnim
        return result
    def __mul__(self,other):
        result=Complex(0,0)
        result.real=self.real*other.real-self.mnim*other.mnim
        result.mnim=self.real*other.mnim+other.real*self.mnim
        return result
    def __truediv__(self,other):
        result=Complex(0,0)
        result.real=(self.real*other.real+self.mnim*other.mnim)/(other.mnim**2+other.real**2)
        result.mnim=(self.mnim*other.real-self.real*other.mnim)/(other.mnim**2+other.real**2)
        return result
    def __abs__(self):
        result=(self.real**2+self.mnim**2)**(1/2)
        return result
    def __str__(self):
        if self.real==0:
            return '{}'.format(float(self.mnim))
        if self.mnim>0:
            return '{}+{}i'.format(float(self.real),float(self.mnim))
        elif self.mnim==0:
            return '{}'.format(float(self.real))
        return '{}{}i'.format(float(self.real),float(self.mnim))
s=Complex(2,3)
print(s)
a=Complex(5,6)
print(a==s)
print(a/s)
print(s/a)
print(abs(a))
````

### Задание 4: Функция возвращает функции! (2 балла)

Отвлечемся немного от классов, вспомним прошлые занятия про функции.  
Реализуйте функцию checker_mod(x, mod=0), возвращающую лямбда функцию от одного аргумента y, которая будет возвращать True, если остаток от деления y на x равен mod, и False иначе.

Пример использования:
```python3
mod_2 = checker_mod(2)

print(mod_2(3)) # False
print(mod_2(4)) # True

mod_2_1 = checker_mod(2, 1)
print(mod_2_1(3)) # True
```

````python
def checker_mod(x,mod=0):
    return lambda y: y%x==mod
mod_2 = checker_mod(2)

print(mod_2(3)) 
print(mod_2(4)) 

mod_2_1 = checker_mod(2, 1)
print(mod_2_1(3)) 
````
